```python
import matplotlib.pyplot as plt
from scipy import stats

data['z_score'] = stats.zscore(data["median"])
z_score_threshold = 2
anomalous_pm25_data = data[abs(data['z_score']) > z_score_threshold]

# Create the figure
plt.figure(figsize=(16, 10), dpi=100)

plt.scatter(data['date'] , data['median'], label='PM2.5 Values', color='green' ,  alpha=0.5)
plt.scatter(anomalous_pm25_data['date'],  anomalous_pm25_data['median'], label='Anomalous PM2.5 Values', color='red'  )
plt.xlabel('Date')
plt.ylabel('PM2.5 Values')
plt.title('PM2.5 Values with Anomalous Data Points in Donetsk from 2020 to 2024', fontsize=30)
plt.legend()
plt.show()


```
